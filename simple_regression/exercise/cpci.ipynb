```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df = df.rename(columns={'per capita income (US$)': 'income_pc'})
df
```

```python
plt.xlabel('year')
plt.ylabel('income per capita (USD)')
plt.scatter(df.year, df.income_pc, color='red', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df.income_pc)
```

```python
reg.predict([[2005]])
```

```python
plt.xlabel('year')
plt.ylabel('income per capita (USD)')
plt.scatter(df.year, df.income_pc, color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
plt.show()
```
